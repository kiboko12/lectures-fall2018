## Probability

A fundamental quantity in statistics is *probability*.  Essentially, a probability is a number between 0 and 1 that describes a degree of belief.  If the probability of a statement $A$ (which we denote $p(A)$) is 1, we have complete certainty that $A$ is true, and if $p(A)=0$, we are certain that $A$ is not true.  Numbers in between 0 and 1 describe a partial degree of certainty -- I might believe there is a $p=0.5$ chance it will rain tomorrow, i.e. whether it rains or not could go either way. Of course, we are 100% certain that it will either rain or it won't, so $p(A)+p(-A) = 1$  If $A$ is instead a continuous variable, e.g. a person's height, we have that $\int p(A) dA = 1$, since we are certain that the person has some height.  In the continuous case $p(A)$ is now the *probabilility density function* (PDF), with units of $1/[A]$.

Another quantity of interest is the *conditional probability*: It's the probability that $A$ is true given some other fixed knowledge $B$. We denote this $p(A|B)$.  In general, $p(A|B)\neq p(A)$, because the truth of $A$ is not independent of the truth of $B$. If I know it's the month of April, my belief about there being raining weather tomorrow is different from if I know that it's the month of July.  

Lastly, we have the *joint probability*: It's the probability that multiple things are true, e.g. that $A$ and $B$ are both true. We denote this probability $p(A,B)$.  __If $A$ and $B$ are *independent*__, $p(A,B)  = p(A)p(B)$. 


## Bayes' Rule

Now we can notice that there is a relation between conditional and joint probabilities.  If we want to know $p(A,B)$, we might be tempted to just compute $p(A)p(B)$, but this is incorrect, because in general $A$ and $B$ are not independent.  Instead of $p(A)$, we need $p(A|B)$, because if $B$ is true, our belief in $A$ changes.  So we have:
$$p(A,B) = p(A|B)p(B)$$
Of course, $p(A,B) = p(B,A)$, so we can do an exchange of $A$ and $B$ and be left with the same probability.  This yields Bayes' Rule:
$$p(A|B)p(B) = p(B|A)p(A)$$.

## Bayesian Inference
Here's general question we can ask in most any scientific context: "How likely is a hypothesis $H$ given that I collected data $D$?  We can straighforwardly write this in our mathematical notation: $p(H|D)$.  Using Bayes' Rule, we find:
$$p(H|D) = \frac{p(D|H)p(H)}{p(D)}$$
Let's pick apart the right-hand-side:
* $p(D|H)$ this is the *likelihood*, which we saw in a previous lecture.  It's how likely we are to get data $D$ given that $H$ is the case.  It's often denoted $\mathcal{L}(H)$ because the data is fixed.
* $p(H)$ is the *prior probability*.  It is how likely we think $H$ is if we have no data at all to guide us.
* $p(D)$ is the *evidence*, i.e. the probability of getting dataset $D$ unconditioned on a specific hypothesis $H$.  In general it is treated as a normalization constant, because the data set is fixed.  If we have a set of possible hypotheses and know that one of them must be the true hypothesis that generated the data, then we get that $p(D) = \int p(D|H)p(H) dH$ to ensure that $p(H|D)$ integrates to 1 over all hypotheses. That is, 
$$1 = \int p(H|D) dH = \int \frac{p(D|H)p(H)}{p(D)} dH = \frac{1}{p(D)}\int p(D|H)p(H) dH \\ \rightarrow p(D) = \int p(D|H)p(H) dH$$

So, if we want to compute the probability of a hypothesis $H$ given some data $D$, we just need to compute the likelihood and choose a prior! For now we don't need to worry about the evidence, since it's just a normalization that we can compute at the very end.

There are a few quantities we often want to compute with posterior distributions, namely the mean (expectation value), median (50th percentile value) and mode (maximum *a posteriori* value).  Let's quickly review them:
* The mean of a random variable $x$ is $x$ integrated against its probability density function.
* The median of random variable $x$ is the value $x_{med}$ for which there is equal total probability above and below $x_{med}$.  
* The mode is the value for which the PDF is largest
  

## Example: The Monty Hall Problem
A classic game-show game once hosted by Monty Hall.  In this game, a contestant must choose one of three doors A, B, and C and wins the prize behind the chosen door.  Behind two doors is a goat, and behind one is a car.  After the contestant has chosen a door, Monty opens one of the other two doors which has a goat behind it to show where one of the goats is.  The contestant now has the opportunity to switch to the other unopened door if they'd like.  Should the contestant switch?

#### Solution
Let's look at this using Bayes' Rule. We'll say that the door originally chosen by the contestant is door A (the letters are just labels of the door, so we can call the door whatever we want).  Monty opens door B to show a goat.  We want to know the probability of the car being in door A versus door C given that door B has a goat behind it. In other words, we need to compute:

$$p(\textrm{A has car}|\textrm{Monty opens B})$$
Using Bayes' Rule:

$$p(\textrm{A has car}|\textrm{Monty opens B}) = p(\textrm{Monty opens B}|\textrm{A has car})p(\textrm{A has car})/p(\textrm{Monty opens B})$$

$p(\textrm{Monty opens B}|\textrm{A has car})$ is 0.5 because if the car is in door A, Monty is free to choose door B or door C to open, since both have goats.  $p(\textrm{A has car})=1/3$, because under no other information, the car could equally likely be behind any of the doors.  $p(\textrm{Monty opens B})$ is a little trickier.  We need to consider all the ways in which you might have chosen your first door, because Monty cannot open the door that you have chosen.  Analogously to computing the evidence integral, we get:

$$p(\textrm{Monty opens B}) = p(\textrm{Monty opens B}|\textrm{chose A})p(\textrm{chose A}) + p(\textrm{Monty opens B}|\textrm{chose B})p(\textrm{chose B}) + p(\textrm{Monty opens B}|\textrm{chose C})p(\textrm{chose C})$$ 

The probability of choosing one of the doors first, say A, is $p(\textrm{chose A}) = 1/3$ since all three doors are equally likely to be chosen by the contestant given no other information. $p(\textrm{Monty opens B}|\textrm{chose B}) = 0$ since Monty can't open door B if it's originally chosen by the contestant. Now $p(\textrm{Monty opens B}|\textrm{chose A})$ is a bit trickier to compute, because we don't know which door the car is actually behind.  Again, like the evidence integral, we can write:

$$p(\textrm{Monty opens B}|\textrm{chose A}) = \sum_{X = A,B,C}p(\textrm{Monty opens B}|\textrm{chose A, car behind X})p(\textrm{car behind X})$$

$p(\textrm{car behind X})$ in all cases is (1/3), so computing this sum, we get $(1/3)*((1/2) + (0) + (1)) = 1/2$.  So now putting everything back together we find that 

$$p(\textrm{A has car}|\textrm{Monty opens B}) = \frac{(1/2)(1/3)}{(1/2)} = \frac{1}{3}$$

This is a surprising result!! It *seems* like once Monty opens door B, the other goat is equally likely to be behind door A or C.  However this intuition turns out to be totally wrong!  Two-thirds of the time one should switch doors after Monty has shown one of the goat doors!  Why does the intuition fail here? Discuss with a partner.

## Multiple Parameters
In many cases, we're not just trying to estimate one value from the data.  Take for example the case of fitting a straight line to some data with $\{(x_i,y_i)\}$ pairs and Gaussian uncertainties on the $y$'s, $\{\sigma_{yi}\}$.  A straight line has two parameters: the slope and the intercept. These two parameters have a joint posterior distribution which we want to estimate. Using Bayes' Rule, it's given by: 

$$p(m,b|\{(x_i,y_i,\sigma_{yi})\}) \sim p\left(\{y_i\}|m,b,\{(x_i,\sigma_{yi})\}\right)p(m,b)$$

Once we have this posterior, a common thing to produce is 1-d posteriors for each parameter, which we call *marginal distributions*.  If we want to just know $p(m|\{(x_i,y_i,\sigma_{yi})\})$, we need to *marginalize out* $b$, which amounts to averaging the joint posterior on $m$ and $b$. In other words, and using Bayes' Rule:

$$p(m|\{(x_i,y_i,\sigma_{yi})\}) = \int p(m|b,\{(x_i,y_i,\sigma_{yi})\})p(b|\{(x_i,y_i,\sigma_{yi})\}) db$$

In this way, one can quote confidence intervals on individual parameters.  This is what we do in our LIGO papers when we report values for e.g. masses and spins of merging black holes we detect.  

## A First Pass at Parameter Estimation

Continuing with line fitting, let's assume that the $\{(x_i,y_i)\}$ data points are generated from some model $y=mx+b$ with $m$ and $b$ as free parameters, but with some Gaussian scatter given by the y-uncertainties, the likelihood is given by:

$$p(\{y_i\}|m,b,\{x_i\},\{\sigma_{yi}\}) = \prod \frac{1}{\sqrt{2\pi \sigma_{yi}}} \exp\left(-\frac{1}{2}\frac{[y_i - mx_i - b]^2}{\sigma_{yi}^2}\right)$$

Let's use this to fit a line to some data to find the slope and intercept that generated the data.  We'll just try to find the maximum a posterior values of the slope and intercept.


```python
from __future__ import print_function
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.integrate import cumtrapz
import scipy.stats as ss
import scipy.optimize as sop
```

```python
# First generate some data from the "truth"
m = 1.
b = 0.
xi = np.linspace(0,10,10)
sig = 0.5
yi = m*xi+b + np.random.normal(0,sig,len(xi))

# define log likelihood:
def line_ll(params,xi,yi,sig):
    m = params[0]
    b = params[1]
    y_model = m*xi+b
    return np.sum(-(1./2)*((y_model - yi)**2.)/(sig**2.) - (1./2)*np.log(2.*np.pi*sig))

# define flat prior on m and b between -10 and 10:
def line_logprior(params):
    if ((params[0]<10) and (params[0]>-10) and (params[1]<10) and (params[1]>-10)):
        return 0
    else:
        return -np.inf

def neg_line_post(params,xi,yi,sig):
    return -(line_ll(params,xi,yi,sig) + line_logprior(params))

result = sop.minimize(neg_line_post,
                      x0=[2,-4], # starting values
                      args=(xi,yi,sig))
m_est,b_est = result['x'][0],result['x'][1]
print('your estimated slope and intercept are:')
print('m = %f, b = %f'%(m_est,b_est))

plt.errorbar(xi,yi,yerr=sig,ls="none",marker='o',label='data')
plt.plot(xi,m_est*xi+b_est,label='model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
```

The fitter seems to work alright! Try messing around with different $\sigma$ values and different numbers of observations and see what happens.  The problem with the above method is that we only get the best fit value.  We can certainly compute the full posterior for any $m$ and $b$ if we want, but usually we don't know the posterior analytically.  This is where sampling comes in.

## Sampling
Often times, one wants obtain samples from a PDF.  It's easier to work with samples than it is to work with than the PDF itself. As an example, suppose the PDF of human heights $p(h)$ is described by a Gaussian.  Suppose now we want to compute the $p(f)$ where $f = f(h)$.  To do that, we need to use the chain rule: 

$$p(f) = \frac{dP}{df} = \frac{dP}{dh}\frac{dh}{df} = p(h)\frac{dh}{df}$$

Note that we've used upper-case $P$ here to denote the cumulative probability $P$ between $f$ and $f+df$.  So it seems that if we want to know $p(f)$, we have to also know $\frac{dh}{df}$ and $p(h)$, but in many practical cases, we can't easily compute $\frac{dh}{df}$ and $p(h)$ may not be an analytically evaluable function.  $f$ might also be computed from some complicated algorithm.  In these cases, it's easier to just apply $f$ to samples from $p(h)$ and treat those transformed samples as samples from $p(f)$.

### Inverse Transform Sampling
The trick to inverse transform sampling is to compute the *cumulative density function* of the PDF, which is defined as $P(x\leq X) = \int_{-\infty}^X p(x)$.  This is just the total probability $P$ that $x \leq X$, and its range is 0 to 1.  To get samples from $p(x)$, sample a random number $u$ from a uniform distribution between 0 and 1, $u\sim U(0,1)$ and apply $P^{-1}(u)$, where $P^{-1}$ is the inverse of $P$.  Let's give it a shot with a Gaussian:

```python
def Gaussian(x):
    return (1./np.sqrt(2.*np.pi))*np.exp(-x**2/2.)   
```

```python
# set up figure
fig, ax = plt.subplots(1,3,figsize=(10,4))

# Plot the Gaussian
x = np.linspace(-10,10,2000)
ax[0].plot(x,Gaussian(x))
ax[0].set_title('Gaussian PDF')
ax[0].set_xlabel('x')
ax[0].set_xlim([-10,10])
ax[0].set_ylim([0,0.5])


# Compute and plot the Gaussian CDF
CDF = cumtrapz(Gaussian(x))*np.diff(x) # int(p(x)dx)
ax[1].plot(x[:-1],CDF)
ax[1].set_title('Gaussian CDF')
ax[1].set_xlabel('x')

# Use an interpolator to compute P^-1 and apply it to U(0,1) samples
u = np.random.uniform(0,1,2000) # random U(0,1) samples
Pinv = np.interp(u,CDF,x[:-1]) # linearly interpolate x(CDF) = P^-1
ax[2].hist(Pinv,bins=25)
ax[2].set_title('Gaussian Samples Histogram')
ax[2].set_xlabel('x')
ax[2].set_xlim([-10,10])
ax[2].set_ylabel('counts')

fig.tight_layout()
```

Pretty darn good!  This method is quick and dirty way to get samples from an arbitrary 1-d distribution.  The place where you need to be careful in this method is in choosing a small enough $dx$ that your CDF values and linear interpolation of CDF$^{-1}$ are accurate.

### Rejection Sampling
Take a look at the Gaussian PDF graph above.  Suppose we were to randomly throw down points anywhere in the area of the graph. Some points would end up below the PDF, some above.  If we throw away all the points above the PDF, the remaining points' x-values will be samples from the PDF.  This is because there is more area for points to be kept at x-values with higher PDF values.  This is a simple way to think about *rejection sampling*.  Let's lay out the simplest version of this algorithm:
* choose a random $x$
* choose a random value $u$ uniformly between 0 and $\max(\textrm{PDF})$
* If $u>\textrm{p}(x)$, don't keep this $x$ sample. Otherwise do keep the sample.

### Markov-Chain Monte-Carlo Sampling

As we mentioned before, it's usually easier to apply any functions of interest to samples rather than trying to analytically deal with PDFs. This technique is known as running a *Monte-Carlo* simulation.  A *Markov Chain* is a sequence of events in which the probability of obtaining an event in the sequence only depends on the last event in the sequence.  A Markov-Chain Monte-Carlo (MCMC) sampler puts these two ideas together to generate samples from a target distribution.  A common MCMC implementation is the Metropolis-Hastings algorithm, which is a form of rejection sampling.  Here's the algorithm:

* Start with an initial sample at $x_i$.
* Use a *proposal distribution* to choose a new point $x_{i+1}$.  Often, the proposal distribution for a new point $x_{i+1}$ is a Gaussian of some width centered at $x_{i}$.
* Calculate the acceptance ratio $\alpha = p(x_{i+1})/p(x_i)$.
* Generate a random number $u$ between 0 and 1, and if $\alpha \geq u$ add $x_{i+i}$ to the chain.  Otherwise, copy $x_i$ and add it to the chain.
* Repeat.

As the chain increases, the samples converge to the target distribution.  Of course, the samples in the chain won't start out even close to the target distribution, so one needs to let the chain run for a while before the samples start to "settle" to equilibrium under the distribution. This is called *burning in* samples, and it amounts to throwing away the first bit of the chain.  A rule of thumb is that the burn-in should go till the acceptance fraction is ~1/4.  Also, each sample is by design correlated with the previous samples,  so one needs to take every $n$-th sample rather than each sample in the chain.  The value of $n$ to choose can be informed by the *autocorrelation length* which tells us how far apart in the chain correlations exist. Lastly, it is worth noting that the best proposal distribution or "jump proposals" are tuneable.  Software like $\texttt{emcee}$ will figure out jump proposals for you, but it's worth knowing that this is something that could affect your results.  Here's a summary of these concepts:

* Burn in: Let the samples settle into an equilibrium under the target distribution.  i.e. throw out the first some-odd samples.
* Acceptance fraction: the fraction of proposals that are accepted
* Autocorrelation Length: The autocorrelation between the values at two different times or positions in an array is:

$$A(\tau) = \frac{\mathbb{E}([x(t)-\mu][x(t+\tau)-\mu])}{\sigma^2}$$

Where $\mu$ and $\sigma^2$ are the mean and variance of $x$.  The easiest way to calculate it is:

$$A(\tau) = \texttt{IFFT}([F[f]F^*[f]])$$

where $F[f]$ is the $\texttt{FFT}$ of x(t).

## A Gravitational-Wave Example

Let's suppose we know there is a gravitational wave in our strain data and we want to estimate the parameters of the source that created the gravitational wave.  This is what we do in LIGO once we have detected that a signal is present in the data.  In probability-speak, we want to know $p(\vec{\theta}|D)$ where $\vec{\theta}$ is the set of model parameters (say, masses and spins of two merging black holes) and $D$ is the strain data from the detector(s). Using Bayes' Rule, we know we need to choose a prior on $\vec{\theta}$ and determine a likelihood function in order for us to get the posterior on $\vec{\theta}$.  The likelihood used in LIGO analyses is:

$$\mathcal{L}(\vec{\theta}) \sim \exp\left[-\frac{1}{2}\langle D-h(\vec{\theta})|D-h(\vec{\theta})\rangle\right]$$

Where $D$ is the strain data recorded and $h(\vec{\theta})$ is the strain predicted by the model.  \langle x | y \rangle$ is the *noise-weighted inner product* defined as:

$$ \langle x | y \rangle = 4\textrm{Re}\int_0^\infty \frac{\tilde{x}(f)\tilde{y}^*(f)}{S(f)}df$$

In words, this likelihood represents the probability of the residual between the data and model ($D-h$) under stationary Gaussian noise described by noise power spectral density (PSD) $S(f)$.  Brushing a number of caveats aside, the PSD can be estimated by taking the amplitude squared of the Fourier transform of the noise.  In LIGO, we have colored noise, meaning that the PSD is not flat as a function of frequency.  

### The Rubber Hits the Road

We'll now use MCMC sampling to get posterior samples of parameters describing a GW we have injected into some noise. In particular, we will try to estimate the distance, chirp mass, and coalescence time of the GW. One last important piece we need is a waveform model.  We'll use the quadrupole formula and Keplerian approximation of circular orbits as our waveform model.  This is given by (see Cutler and Flanagan 1994):

$$ \tilde{h}(f) = G^{5/6}c^{-3/2}\frac{Q}{d_L}\mathcal{M}^{5/6}f^{-7/6}\exp[i\Psi(f)]$$

where

$$ \Psi(f) = 2\pi f t_c - \phi_c -\frac{\pi}{4} + \frac{3}{4}(8\pi\frac{G}{c^3} \mathcal{M}f)^{-5/3}.$$

Note that this waveform is constructed in the frequency domain, not the time domain. Let's code it up and leave three open parameters:  The chirp mass $\mathcal{M}$, the distance $d_L$, and the coalescence time $t_c$.  We'll fix other values like $Q$ (orientation of detectors and binary) and phase at coalescence $\phi_c$.

```python
# define constants
c = 3e8 # m/s
G = 6.67408e-11 # m^3 kg^-1 s^-1
Msol = 2e30 # kg
Mpc = 3e22 # m

# define waveform functions
def h_tild(f,d_L,Mc,t_c):
    return G**(5./6) * c**(-3./2) * Mc**(5./6) * f**(-7./6)*np.exp(1j*Psi(f,t_c,Mc)) / d_L

def Psi(f,t_c,Mc):
    return 2*np.pi*f*t_c - np.pi/4 + (3./4)*(8*np.pi*(G/(c**3))*Mc*f)**(-5./3)

# set some nominal parameters for a test
n_s = 8192
f_max = 300.
f = np.linspace(0,f_max,n_s) #number of samples
fcalc = f[f>=30]
below_cutoff = np.sum(f<10)
d_L = 200.*Mpc
Mc = 50.* Msol
t_c = 0

#plot
fig,ax = plt.subplots(1,3,figsize=(11,3))
ax[0].loglog(fcalc,np.abs(h_tild(fcalc,d_L,Mc,t_c)))
ax[0].set_xlabel('Frequency (Hz)')
ax[0].set_ylabel('Amplitude')


ax[1].semilogx(fcalc,np.unwrap(np.angle(h_tild(fcalc,d_L,Mc,t_c))))
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Phase')

# make a new h_tild that includes frequencies down to zero
h_tild_new = np.append(np.zeros(below_cutoff),h_tild(fcalc,d_L,Mc,t_c))[::-1]
# window
from scipy.signal import tukey
window = tukey(len(h_tild_new),alpha=0.5)
# take the inverse FFT to go back to the time domain
h = np.fft.irfft(h_tild_new*window)


ax[2].plot(np.linspace(0,n_s/(2.*f_max),n_s),h[-n_s:])
ax[2].set_xlabel('time')
ax[2].set_ylabel('strain')

fig.tight_layout()

```

Great, now that we have a waveform, let's do some parameter estimation.  We'll use white noise here, but in general we have colored noise in real interferometers.  First let's define our likelihood

```python
def loglike(params,data,noise_level):
    d_L,Mc,t_c = params[0]*Mpc,params[1]*Msol,params[2]
    h_pred = h_tild(fcalc,d_L,Mc,t_c)
    res = data - h_pred
    res_sqr = 4.*np.square(np.abs(res))
    val =  -(1./2)*np.sum(res_sqr*np.diff(fcalc)[0])/(noise_level**2)
    if np.isnan(val):
        return -np.inf
    else:
        return val

def logprior(params):
    d_L,Mc,t_c = params[0],params[1],params[2]
    if (((d_L > 500) or (d_L < 0)) or ((Mc > 55) or (Mc < 45)) or ((t_c > 0.005) or (t_c < -0.005))):
        return -np.inf
    else:
        return np.log(d_L**2.)

def logpost(params,data,noise_power):
    return loglike(params,data,noise_power) + logprior(params)
```

```python
# create real signal
d_L_Mpc = 321.6 
d_L = d_L_Mpc * Mpc
Mc_Msol = 52.4
Mc = Mc_Msol * Msol
t_c = 0.0 # s

h_tild_real = h_tild(fcalc,d_L,Mc,t_c)

# create noise at, say, the 1e-22 level
noise_level = 1e-22
noise = np.random.normal(0,noise_level,len(h_tild_real))
noise = noise*np.exp(1j*np.random.uniform(0,2.*np.pi,len(noise)))

# put the signal in the noise
h_noisy = h_tild_real + noise

fig,ax = plt.subplots(1,3,figsize=(11,3))
ax[0].loglog(fcalc,np.abs(h_noisy))
ax[0].set_xlabel('Frequency (Hz)')
ax[0].set_ylabel('Amplitude')


ax[1].semilogx(fcalc,np.unwrap(np.angle(h_noisy)))
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Phase')

```

```python
fig,ax = plt.subplots(1,3,figsize=(11,3))

for mci in np.linspace(40,60,10):
    h = h_tild(fcalc,d_L,mci*Msol,t_c)
    ax[0].loglog(fcalc,np.abs(h))
    ax[0].set_xlabel('Frequency (Hz)')
    ax[0].set_ylabel('Amplitude')


    ax[1].semilogx(fcalc,np.unwrap(np.angle(h)))
    ax[1].set_xlabel('Frequency (Hz)')
    ax[1].set_ylabel('Phase')
    print(mci,loglike([d_L_Mpc,mci,t_c],h_noisy,noise_level))
print(loglike([d_L_Mpc,Mc_Msol,t_c],h_noisy,noise_level))
```

```python
import emcee

ndim, nwalkers = 3, 20
p0 = [np.array([np.random.normal(d_L_Mpc,d_L_Mpc*0.1),np.random.normal(Mc_Msol,Mc_Msol*0.1),np.random.normal(t_c,0.01)]) for i in range(nwalkers)]

sampler = emcee.EnsembleSampler(nwalkers, ndim, logpost, args=[h_noisy,noise_power])
sampler.run_mcmc(p0, 1000);
```
